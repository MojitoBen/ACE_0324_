```python
import cv2

cap = cv2.VideoCapture('/home/asc/鋼卷標籤/2022-10-04/001/dav/12/12.00.00-12.10.00[R][0@0][0].dav')
count=0
while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Can't receive frame (stream end?). Exiting ...")
            break
        print(frame.shape)
        if count%5==0:
            cv2.imwrite('/home/asc/鋼卷標籤/img/10_03/%05d'%count+'.png',frame)
        
        count+=1
        cv2.imshow('eee',frame)
        if cv2.waitKey(1) == ord('q'):
            break
cv2.destroyAllWindows()
cap.release()
```
